# Der Boxplot

```python
# Import von Matplotlib's Pyplot-Funktionen
import matplotlib.pyplot as plt

# Daten der Wissenschaftler (Messungen der Körpergröße)
data = {
    'Dorftown': [160, 165, 170, 175, 180, 185, 190, 215],
    'Stadtort': [150, 155, 160, 165, 170, 175, 180, 200],
    'Megacity': [140, 150, 160, 170, 180, 190, 200, 210]
}

# Boxplot erstellen
plt.figure(figsize=(8, 6))
plt.boxplot(data.values(), labels=data.keys())
plt.title('Boxplot der Körpergrößen je Stadt')
plt.ylabel('Körpergröße in cm')
plt.grid(True)

# Boxplot zeigen
plt.show()
```
