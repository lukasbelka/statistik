```python
import matplotlib.pyplot as plt

# Daten
linksschief = [2, 3, 4, 5, 9, 13, 19, 24, 32, 40, 8, 5]
rechtsschief = [4, 7, 40, 32, 24, 19, 13, 9, 5, 4, 3,2]

# Balkendiagramm linkssschief
plt.figure(figsize=(8, 5))
plt.bar(range(len(linksschief)), linksschief, color='skyblue')
plt.title('Linksschiefe Verteilung', fontsize=14)
plt.xlabel('Beobachtung', fontsize=12)
plt.ylabel('Wert', fontsize=12)
plt.xticks(range(len(linksschief)), labels=[str(i) for i in range(len(linksschief))])  # Benutzerdefinierte tick labels auf der x-Achse
plt.grid(axis='y')
plt.show()

# Belkendiagramm rechtsschief
plt.figure(figsize=(8, 5))
plt.bar(range(len(rechtsschief)), rechtsschief, color='skyblue')
plt.title('Rechtsschiefe Verteilung', fontsize=14)
plt.xlabel('Beobachtung', fontsize=12)
plt.ylabel('Wert', fontsize=12)
plt.xticks(range(len(rechtsschief)), labels=[str(i) for i in range(len(rechtsschief))])
plt.grid(axis='y')
plt.show()
```
